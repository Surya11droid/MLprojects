```python
x= df.drop(columns= ['price'], axis=1)
y= df['price']
x
```

One Hot Encoding

```python
x_one_encode= pd.get_dummies(x, columns=['model', 'transmission', 'fuelType'], drop_first= True)
x_one_encode= x_one_encode.astype(int)
x_one_encode.columns
```

Label Encoding

```python
from sklearn.preprocessing import LabelEncoder



columns= ['model','transmission','fuelType']
xlabel= x
for i in columns:
    encode= LabelEncoder()
    xlabel[i]= encode.fit_transform(xlabel[i])

```

```python
x_one_encode
```

```python
x_one_encode.columns
```

Standard scaling

```python
from sklearn.preprocessing import StandardScaler
cols = ['year', 'mileage', 'tax', 'mpg', 'engineSize']
scalar = StandardScaler()
x_one_encode[cols] = scalar.fit_transform(x_one_encode[cols])
x_one_encode
```

```python
xlabel.columns
```

```python
xlabel[['model', 'year', 'transmission', 'mileage', 'fuelType', 'tax', 'mpg',
       'engineSize']]= scalar.fit_transform(xlabel[['model', 'year', 'transmission', 'mileage', 'fuelType', 'tax', 'mpg',
       'engineSize']])
xlabel
```

```python
from scipy.stats import pearsonr
#pearson correlation calculation
selected_features= ['model', 'year', 'transmission', 'mileage', 'fuelType', 'tax',
       'mpg', 'engineSize']
corelations = {feature: pearsonr(x[feature], df['price'])[0] for feature in selected_features}
corelations_df = pd.DataFrame(list(corelations.items()), columns=['Feature', 'Correlation'])
corelations_df= corelations_df.sort_values(by='Correlation', ascending=False)
print(corelations_df)
```

one hot encoding model

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error,r2_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(x_one_encode, y, test_size=0.2, random_state=42)  
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred= model.predict(X_test)
y_pred
```

```python
r2= r2_score(y_test, y_pred)
print(f'R2 Score: {r2}')
n= X_test.shape[0]
p= X_test.shape[1]
ADJUSTED_R2 = 1 - (1 - r2) * (n - 1) / (n - p - 1)
print(f'Adjusted R2 Score: {ADJUSTED_R2}')
```
